```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.chrome.service import Service as ChromeService
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

import time
import re
```

```python
url = "https://www.baseball-almanac.com/hitting/hihr1.shtml"
```

```python
options = webdriver.ChromeOptions()
options.page_load_strategy = "none"
driver = webdriver.Chrome(options=options, service=ChromeService(ChromeDriverManager().install()))
wait = WebDriverWait(driver, 20)

driver.get(url)

wait.until(EC.presence_of_element_located((By.CLASS_NAME, 'boxed')))
original_tab = driver.current_window_handle

driver.execute_script("window.stop();") #alert('Content is located, loading stopped!')")
```

```python
table = driver.find_element(By.CLASS_NAME, 'boxed')
```

```python
players = driver.find_elements(By.XPATH, ".//a[contains(@href, 'players/player')]")
home_runs = driver.find_elements(By.XPATH, ".//a[contains(@href, 'players/home_run')]")
```

```python
player_names = []
player_links = []
player_ids = []
home_runs_totals = []
quotes = []
bat_hands = []
colleges = []

for i in range(1001):
    player = driver.find_elements(By.XPATH, ".//a[contains(@href, 'players/player')]")[i]
    home_run = driver.find_elements(By.XPATH, ".//a[contains(@href, 'players/home_run')]")[i]
    player_names.append(player.text)
    player_links.append(player.get_attribute("href"))
    player_ids.append(re.search(r'(?<=\?p=)([a-z0-9]+)', player.get_attribute("href")).group())
    home_runs_totals.append(int(home_run.text))
    # player.click()
    # quote = driver.find_element(By.XPATH, ".//div[contains(@class, 'topquote')]").text
    # info = driver.find_elements(By.XPATH, ".//td[contains(@class, 'biocolpad')]")
    # bat_hand = info[19].text
    # college = info[17].text
    # quotes.append(quote)
    # bat_hands.append(bat_hand)
    # colleges.append(college)
    # driver.back()
    
    print("Back")

# for (player, home_run) in zip(players, home_runs):
#     print(player)
#     player_names.append(player.text)
#     home_runs_totals.append(int(home_run.text))
#     player.click()
#     driver.back()
```

```python
for url in player_links:
    driver.switch_to.new_window("tab")
    driver.get(url)
    wait.until(EC.presence_of_element_located((By.XPATH, ".//td[contains(@class, 'biocolpad')]")))
    driver.execute_script("window.stop();")
    quote = driver.find_element(By.XPATH, ".//div[contains(@class, 'topquote')]").text
    info = driver.find_elements(By.XPATH, ".//td[contains(@class, 'biocolpad')]")
    bat_hand = info[19].text
    college = info[17].text
    quotes.append(quote)
    bat_hands.append(bat_hand.strip())
    colleges.append(college.strip())
    driver.close()
    driver.switch_to.window(original_tab)
    
```
